```python
from IPython.core.display import HTML
from bs4 import BeautifulSoup, SoupStrainer
from IPython.display import IFrame
import urllib # package required to interact with live webpage
import pandas as pd # will use to store the data from the webpage
import re
import requests
import time
```

```python
url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
html = urllib.request.urlopen(url).read().decode('utf-8')
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
links = soup.find_all('a')
```

```python
ratings = soup.find_all('td', {'class': 'ratingColumn imdbRating'}) 
```

```python
start = time.process_time()
movies = {}
i = 0
for link in links:
    if '/" title=' in str(link):
        key = i
        title = (re.search('>(.*)<', str(link))).group(1)
        director = (re.search('title="(.*) \(dir.\)', str(link))).group(1)
        cast = (re.search('\(dir.\), (.*)">', str(link))).group(1)
        rating = (re.search('title="(.*) based', str(ratings[i]))).group(1)
        
        movielink = "https://www.imdb.com" + (re.search('href="(.*)" title', str(link))).group(1)
        moviehtml = requests.get(movielink)
        moviesoup = BeautifulSoup(moviehtml.content, "lxml", parse_only = SoupStrainer(
                       'a', class_ = 'sc-16ede01-3 bYNgQ ipc-chip ipc-chip--on-baseAlt'))
        elements = moviesoup.find_all('a', {'class': 'sc-16ede01-3 bYNgQ ipc-chip ipc-chip--on-baseAlt'})
        genres = [element.text for element in elements]
        
        movies[key] = {"title":title, "director":director, "cast":cast, "rating": float(rating), "genres" : genres}
        i += 1
print(time.process_time() - start)
```
